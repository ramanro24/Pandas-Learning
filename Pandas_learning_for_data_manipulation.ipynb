Grouping & Advanced Operations
1. Find the total salary paid in each department.
2. Calculate the minimum and maximum salary in each department.
3. Find the median salary for each department.
4. Count how many employees have an age above 30 in each department.
5. Create a pivot table showing average salary by "Department" and "Age".

```python
#1. Find the total salary paid in each department.

df.groupby('Department')['Salary'].sum()

```

```python
#2. Calculate the minimum and maximum salary in each department.
df.groupby('Department')['Salary'].agg(['min','max'])
```

```python
#4. Count how many employees have an age above 30 in each department.

df.groupby('Department')['Age'].apply(lambda x: (x>30).sum())
df.groupby('Department')['Age'].apply(lambda x: (x>40).sum())
```

```python
#5.Calculate average salary by "Department" and "Age".

df.groupby(['Department','Age'])["Salary"].mean()
```
