# Different way of creating dataframe


```python

# Method1:  First, let's import pandas and other useful libraries
import pandas as pd
import numpy as np

# Section 1: Creating DataFrames
# Method 1: From a dictionary
data = {
    'Name': ['John', 'Alice', 'Bob', 'Emma'],
    'Age': [25, 30, 35, 28],
    'Salary': [50000, 60000, 75000, 65000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing']
}
df = pd.DataFrame(data)
df
```

```python
#Method 2: From list of Dictonaries:
data2= [
    {"Name":"Jonny","Age": 23 , "Salary": 3000  , "Department":"HR" },
    {"Name":"Alice","Age": 30 , "Salary": 5000, "Department":"IT"},
    {"Name":"Bob","Age" : 31 , "Salary": 5010, "Department":"IT"},
    {"Name":"Emma","Age" : 22 , "Salary": 1500, "Department": "HR"},

]

df2=pd.DataFrame(data2)
df2
```

```python
# Section 2: Basic DataFrame Operations
# Selecting columns
names=df["Name"]
multiple_coulumns=df[["Name","Age", "Salary"]]
print(names)
print(multiple_coulumns)
```

```python
#Adding new columns:
df["Bonus"] = df["Salary"] *0.1
print(df)
```

```python
#Filtering Data:
high_salary=df[df["Salary"]> 2000]
print(high_salary)
```

```python
#Sorting:
df.sort_values('Age')
```

```python
#Section 3: Data Manipulations:
# Group by department and claucualte the mean of Salary:

department_group=df.groupby('Department')['Salary'].mean()
print(department_group)
```

```python
#Section 4: Apply function to columns:

df['Name_length']=df['Name'].apply(len)
print(df)
```

Create a sales DataFrame with:
- Product
- Price
- Quantity
- Date
Then:


a) Calculate total revenue per product (Price * Quantity)

b) Find top 3 selling products by quantity

c) Calculate monthly sales totals (assuming multiple months of data)

"""

```python
data= {
    "Product":["Rice", "Pulses", "Wheat", "Produce", "Meat", "Frozen"],
    "Price":[22, 21, 34, 21, 48, 21],
    "Quantity":[4, 2, 3, 4, 2, 4],
    "Date":["30-Jan-2024", "20-Feb-2024", "15-March-2024", "16-March-2024", "17-April-2024", "30-April-2024"]
}
df=pd.DataFrame(data)
print(df)
```

```python
#a) Calculate total revenue per product (Price * Quantity)
df['Total Revenue']= df['Price']* df["Quantity"]
print(df)
```

```python
#b) Find top 3 selling products by quantity
top_3_products=df.sort_values('Quantity', ascending=False)
print(top_3_products.head(3))
```

```python
#c) Calculate monthly sales totals (assuming multiple months of data)
df['Date']=pd.to_datetime(df['Date'], format='%d-%B-%Y', errors='coerce')
monthly_sales=df.groupby(df['Date'].dt.to_period('M'))['Total Revenue'].sum()
print(monthly_sales)

```

**Practice Question 4**

Create a DataFrame for an online stores product reviews with:**

- Review_ID (e.g., R001, R002, etc.)

- Product_Name

- Rating (1 to 5)

- Review_Date (dates in 2025)

- Customer_Type (e.g., 'New', 'Returning')

Then:

a) Find products with an average rating below 3

b) Calculate the number of reviews per Customer_Type

c) Add a 'Rating_Category' column:
   - 'Excellent' for Rating = 5
   - 'Good' for Rating = 4
   - 'Average' for Rating = 3
   - 'Poor' for Rating <= 2

```python
data1={
    "Review_ID":['R001', 'R002', 'R003', 'R004', 'R005', 'R006'],
    "Product_Name":[  'Rice', 'Pulses', 'Wheat', 'Produce', 'Meat', 'Frozen'],
    'Reveiew_Date':['April 2024', 'March 2024', 'August 2024', "July 2024", 'June 2024', 'Feb 2024'],
    'Customer_Type':['Non_Returning', 'Returning', 'Returning', 'Non_Returning', 'Non_Returning', 'Non_Returning'],
    'Ratings': [1,5,6,3,2,2]

}
df=pd.DataFrame(data1)
print(df)
```

```python
#1. Find a prodcut where average rating is below 3
average_rating=df.groupby('Product_Name')['Ratings'].mean()[lambda x: x<3]

print(average_rating)
```

```python
#2.  Calculate the number of review per  customer type
review_per_customer=df.groupby('Customer_Type')['Review_ID'].count()
print(review_per_customer)
```

```python
#3. Add rating columns:
df['Ratings_Category']=df['Ratings'].apply(lambda x: 'Excellent' if x==5 else ('Good' if x==4 else ('Average' if x==3 else 'Poor')))
print(df)
```
