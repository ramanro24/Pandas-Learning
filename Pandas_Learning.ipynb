```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert order_date to datetime if not already
data['order_date'] = pd.to_datetime(data['order_date'])

# Extract the year and month as datetime format
data['order_month'] = data['order_date'].dt.to_period('M')

# Group by order_month and sum total revenue
revenue_per_month = data.groupby('order_month')['total_price'].sum().reset_index()

# Convert to datetime and extract full month name
revenue_per_month['order_month'] = revenue_per_month['order_month'].dt.strftime('%B %Y')

# Plot the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(x='order_month', y='total_price', data=revenue_per_month, marker='o', linewidth=2)

plt.xlabel("Order Month")
plt.ylabel("Total Revenue")
plt.title("Total Revenue Per Month")
plt.xticks(rotation=45)  # Rotate for better readability

plt.show()
```

```python
# 7.    Find the total quantity of pizzas ordered in each month.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert order_date to monthly periods
data['order_month'] = pd.to_datetime(data['order_date']).dt.to_period('M')

# Aggregate quantity per month
quantity_per_month = data.groupby('order_month')['quantity'].sum().reset_index()

# Sort by order_month (time-based sorting)
quantity_per_month = quantity_per_month.sort_values(by='order_month')

# Plot the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(x=quantity_per_month['order_month'].astype(str),  # Convert period to string for plotting
             y=quantity_per_month['quantity'],
             marker='o', linewidth=2)

plt.xlabel("Order Month")
plt.ylabel("Total Quantity")
plt.title("Total Quantity Per Month")
plt.xticks(rotation=45)  # Rotate for better readability
plt.grid(True)  # Optional: Add grid for better visualization

plt.show()


```

```python
# 8.	Calculate the percentage contribution of each pizza category to the total sales.
# Group the pizza cateorgy and sum the total sales:
data['Sales'] =  data['quantity'] * data['total_price']
category_sales=  data.groupby('pizza_category')['Sales']. sum().reset_index()

#Percentage contribution of category sales:

category_sales['percentage_contribution'] =  (category_sales['Sales']/category_sales['Sales'].sum()) *100

category_sales
```

```python
category_sales.sort_values(by='percentage_contribution', ascending = False)
plt.figure(figsize = (10,6))
sns.barplot(x = 'pizza_category', y= 'percentage_contribution', data = category_sales, palette= 'viridis')
plt.xlabe("pizza category")
plt.ylabel('Percentage contribution (%)')
plt.xticks(rotation=45)  # Rotate category names for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort the data for better visualization
category_sales = category_sales.sort_values(by='percentage_contribution', ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='pizza_category', y='percentage_contribution', data=category_sales, palette='viridis')

# Labels and title
plt.xlabel("Pizza Category")
plt.ylabel("Percentage Contribution (%)")
plt.title("Percentage Contribution of Each Pizza Category to Total Sales")
plt.xticks(rotation=45)  # Rotate category names for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Ensure order_time is in datetime format, then extract the hour
data['order_time'] = pd.to_datetime(data['order_time'], format='%H:%M:%S').dt.hour

# Count total orders for each pizza category
total_category_counts = data['pizza_category'].value_counts()

# Filter for night orders (after 6 PM)
night_orders = data[data['order_time'] >= 18]

# Count night orders for each pizza category
night_category_counts = night_orders['pizza_category'].value_counts()

# Combine both counts into a DataFrame
category_counts = pd.DataFrame({'Total Orders': total_category_counts, 'Night Orders': night_category_counts}).fillna(0).astype(int)

# Print the result
print(category_counts)
```
